```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
```

```python
urls = [
    "http://lotteriamongolia.mn/%D0%B1%D2%AF%D1%80%D0%B3%D1%8D%D1%80/",
    "http://lotteriamongolia.mn/%D1%82%D0%B0%D1%85%D0%B8%D0%B0/",
    "http://lotteriamongolia.mn/amttan/",
    "http://lotteriamongolia.mn/undaa/"
]
```

```python
product_names = []
product_prices = []
```

```python
for url in urls:

    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    names = []
    prices = []

    products = soup.find_all("h3", {"class": "portfolio-title"})
    for p in products:
        names.append(p.text) 

    prices_elm = soup.find_all("span", {"class": "tw-custom-label"})
    for p in prices_elm:
        prices.append(p.text)

    product_names.extend(names)
    product_prices.extend(prices)

df = pd.DataFrame({
    "Product Names": product_names,
    "Product Prices": product_prices  
})
```
